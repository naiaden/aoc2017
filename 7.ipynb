```python
class Trees:
    def __init__(self, _input):
        self.trees = {}
        self.read_programs(_input)

    def follow_back(self, tree_name):
        t = self.trees[tree_name]
        while t.parent is not None:
            t = self.trees[t.parent]
        return t
        
    def read_programs(self, _input):
        for line in _input:
            parts = line.split(" -> ")
            head_part = parts[0].split(" ")
            self.trees[head_part[0]] = Tree(head_part[0], head_part[1].replace("(", "").replace(")", ""))

        for line in _input:
            parts = line.split(" -> ")
            head_part = parts[0].split(" ")

            if len(parts) > 1:
                for child in parts[1].split(", "):
                    c = self.trees[child]
                    c.parent = head_part[0]
                    self.trees[head_part[0]].add_child(c)

class Tree:
    def __init__(self, head, weight, children=None):
        self.head = head
        self.parent = None
        self.weight = int(weight)
        self.children = []
        self.total_weight = 0
    
    def add_child(self, child):
        self.children.append(child)
    
    def get_weights(self):
        if self.total_weight == 0:
            self.total_weight = self.weight
            if self.children:
                ws = []
                
                w = None
                for child in self.children:
                    cw = child.get_weights()
                    
                    if w is None:
                        w = cw
                    
                    self.total_weight += cw
                    ws.append(cw)
                print(f"[{self.head}] {ws}")
                if len(set(ws)) > 1:
                    w1 = list(set(ws))[0]
                    w2 = list(set(ws))[1]
                    dw = abs(w1-w2)
                    wn = [x for x,y in zip(self.children, ws) if y == max(w1,w2)][0]
#                     print(f">>>>\t[{self.head}] difference: {w1} and {w2} is {dw} --> s.t. {wn.weight} - {dw} = {wn.weight-dw} is the new weight")
                    raise Exception(f">>>>\t[{self.head}] difference: {w1} and {w2} is {dw} --> s.t. {wn.weight} - {dw} = {wn.weight-dw} is the new weight")
        return self.total_weight
    
    def __str__(self):
        return self.to_string(rec=True)
    
    def to_string(self, rec=True):
        if rec:
            return f"[{self.head}:{self.weight}/{self.total_weight}] ({[x.to_string(rec=False) for x in self.children]}))"
        else:
            return f"[{self.head}:{self.weight}/{self.total_weight}]"
```
