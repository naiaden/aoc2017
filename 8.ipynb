```python
class Operator:
    def __init__(self, operator):
        self.operator = operator
    
    def evaluate(self, lhs, rhs):
        if self.operator == ">":
            return lhs > rhs
        elif self.operator == ">=":
            return lhs >= rhs
        elif self.operator == "<":
            return lhs < rhs
        elif self.operator == "<=":
            return lhs <= rhs
        elif self.operator == "==":
            return lhs == rhs
        elif self.operator == "!=":
            return lhs != rhs
        
        if self.operator == "inc":
            return lhs + rhs
        elif self.operator == "dec":
            return lhs - rhs

    def __str__(self):
        return self.operator
    
class Operation:
    def __init__(self, _input):
        self.register, self.operand, self.amount, _, self.cond_left, self.cond_op, self.cond_right = _input.split()
        
        self.operand = Operator(self.operand)
        self.cond_op = Operator(self.cond_op)
        
        self.amount = int(self.amount)
        self.cond_right = int(self.cond_right)
    
    def evaluate(self, state):
        if self.cond_op.evaluate(state.registers[self.cond_left], self.cond_right):
            state.registers[self.register] = self.operand.evaluate(state.registers[self.register], self.amount)

from collections import defaultdict
    
class State:
    def __init__(self):
        self.registers = defaultdict(int)

    def __str__(self):
        return str(self.registers)
    
    def get_largest_value(self):
        return max(self.registers.items(), key=operator.itemgetter(1))[0]
```
